```python
import numpy as np
import cv2
import glob
import matplotlib.pyplot as plt
import pickle
import os
%matplotlib inline
```

```python
# Load undistortion matrix
with open('./camera_cal/camera_calibration_mat.p', mode='rb') as f:
    dist_pickle = pickle.load(f)
    mtx = dist_pickle["mtx"]
    dist = dist_pickle["dist"]
```

```python
# Load test images
images = glob.glob('./test_images/*.jpg')
```

```python
def show_images(images, cols = 4, rows = 5, figsize=(15,10), cmap = None):
    """
    Display `images`
    """
    imgLength = len(images)
    fig, axes = plt.subplots(rows, cols, figsize=figsize)
    indexes = range(cols * rows)
    for ax, index in zip(axes.flat, indexes):
        if index < imgLength:
            img = images[index][1]
            image_name = images[index][0]
            if cmap == None:
                ax.imshow(img)
            else:
                ax.imshow(img, cmap=cmap)
            ax.set_title(image_name)
            ax.axis('off')

images_show = []
for fname in images:
    img_name = os.path.split(fname)[1]
    img = cv2.cvtColor(cv2.imread(fname), cv2.COLOR_BGR2RGB)
    images_show.append((img_name, img))
show_images(images_show, 2, 4, (15, 13))
```

```python
# Undistortion and convert to HLS
img = cv2.imread(images[2])
undist = cv2.undistort(img, mtx, dist, None, mtx)
hls = cv2.cvtColor(undist, cv2.COLOR_RGB2HLS)
# Show different channels of HLS
fig, axes = plt.subplots(ncols=3, figsize=(20,10))
for index, a in enumerate(axes):
    a.imshow(hls[:,:,index], cmap='gray')
    a.axis('off')
```

### Show HLS color space third channel

```python
def undistort(img, mtx=mtx, dist=dist):
    return cv2.undistort(img, mtx, dist, None, mtx)
def to_hls(img, channel=2):
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    return hls[:,:,channel]

images_show_colorspace = []
for item in images_show:
    images_show_colorspace.append((item[0], to_hls(undistort(item[1]))))
show_images(images_show_colorspace, 2, 4, (15, 13), cmap='gray')
```

### Sobel

```python
def binary(img, thresh=(0, 255)):
    xbinary = np.zeros_like(img)
    xbinary[(img >= thresh[0]) & (img <= thresh[1])] = 1
    return xbinary
    
def abs_sobel_thresh(img, orient='x', sobel_kernel=3, thresh=(0, 255)):
    if orient == 'x':
        yorder = 0
        xorder = 1
    else:
        yorder = 1
        xorder = 0
        
    sobel = cv2.Sobel(img, cv2.CV_64F, xorder, yorder, ksize=sobel_kernel)
    abs_sobel = np.absolute(sobel)
    scaled = np.uint8(255.0*abs_sobel/np.max(abs_sobel))
    return binary(scaled, thresh)
```

```python
images_show_sobelx = []
for item in images_show_colorspace:
    images_show_sobelx.append((item[0], abs_sobel_thresh(item[1], thresh = (10, 160))))
show_images(images_show_sobelx, 2, 4, (15, 13), cmap='gray')
```

```python
images_show_sobely = []
for item in images_show_colorspace:
    images_show_sobely.append((item[0], abs_sobel_thresh(item[1], orient='y', thresh = (10, 160))))
show_images(images_show_sobely, 2, 4, (15, 13), cmap='gray')
```

### magnitude of gradient

```python
def mag_thresh(img, sobel_kernel=3, mag_thresh=(0, 255)):
    # Calculate gradient magnitude
    sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
    gradmag = np.sqrt(sobelx**2 + sobely**2)
    scale_factor = np.max(gradmag)/255 
    gradmag = (gradmag/scale_factor).astype(np.uint8) 
    mag_binary = binary(gradmag, mag_thresh)
    return mag_binary
```

```python
images_show_sobelmag = []
for item in images_show_colorspace:
    images_show_sobelmag.append((item[0], mag_thresh(item[1], mag_thresh = (5, 160))))
show_images(images_show_sobelmag, 2, 4, (15, 13), cmap='gray')
```

### Direction of Gradient

```python
def dir_threshold(img, sobel_kernel=3, thresh=(0, np.pi/2)):
    # Calculate gradient direction
    # Apply threshold
    sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
    absgraddir = np.arctan2(np.absolute(sobely), np.absolute(sobelx))
    dir_binary = binary(absgraddir, thresh)
    return dir_binary
```

```python
images_show_sobeldir = []
for item in images_show_colorspace:
    images_show_sobeldir.append((item[0], dir_threshold(item[1], thresh = (0.79, 1.20))))
show_images(images_show_sobeldir, 2, 4, (15, 13), cmap='gray')
```

### Combination

```python
def combineGradients(sobel_x, sobel_y, sobel_mag, sobel_dir):
    """
    Compute the combination of Sobel X and Sobel Y or Magnitude and Direction
    """
    combined = np.zeros_like(sobel_x) 
    # combined[((sobel_x == 1) & (sobel_y == 1)) | ((sobel_mag == 1) & (sobel_dir == 1))] = 1
    combined[(sobel_x == 1) & (sobel_y == 1)] = 1     ## better than above
    return combined
```

```python
images_show_combination = []
for i in range(len(images_show_colorspace)):
    fname = images_show_colorspace[i][0]
    sobel_x = images_show_sobelx[i][1]
    sobel_y = images_show_sobely[i][1]
    sobel_mag = images_show_sobelmag[i][1]
    sobel_dir = images_show_sobeldir[i][1]
    sobel_combined = combineGradients(sobel_x, sobel_y, sobel_mag, sobel_dir)
    images_show_combination.append((fname, sobel_combined))
show_images(images_show_combination, 2, 4, (15, 13), cmap='gray')
```
